```python
import jax
from stochastic_interpolant.UNet import UNet
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python
init_key = jax.random.PRNGKey(1829)
model = UNet(
    key=init_key,
    data_shape=(3,256,256),
    is_biggan=False,
    dim_mults=[1, 2, 4],
    hidden_size=64,
    heads=4,
    dim_head=32,
    dropout_rate=0.0,
    num_res_blocks=2,
    attn_resolutions=[16],
    t1=1,
    langevin=False,
)
```

```python
example = jax.random.normal(jax.random.PRNGKey(1239),(3,256,256))
```

```python
plt.imshow(model(0.3,example).transpose(1,2,0))
```
