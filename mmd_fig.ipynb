```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
examples_to_run = [
    'swissroll',
    'circles',
    'rings',
    'moons',
    '8gaussians',
    'pinwheel',
    '2spirals',
    'checkerboard',
    'line',
    'cos'
]
mmd_results = {}
for ex in examples_to_run:
    mmd_results[ex]=(np.load(f'all_figures/{ex}/mmd.npy'))
```

```python
plt.figure(figsize=(12,5))
plt.plot([1e-3,30],[1,1],c='black')
for name,results in mmd_results.items():
    plot_mmd = np.sqrt(results[:,1]/results[0,1])#np.sqrt(results[:,1]/results[:,0])
    plt.plot(results[:,0],plot_mmd,label = name)
#plt.title("MMD($\epsilon$)/MMD(0)")
plt.ylabel("MMD($\epsilon$)/MMD(0)")
plt.xlabel("$\epsilon$")
plt.xscale('log')
plt.ylim(0.4,1.1)
plt.legend()
plt.savefig("MMD_stochastic.png")
```
